```python
from env import DATA_DIR
from src import KeywordExtractor
from src.utils import load_jsons_data
from src.df_filters import DFFilters
import pandas as pd
from pprint import pprint
```

```python
keyword_extractor = KeywordExtractor()
filter = DFFilters()
```

```python
df = load_jsons_data(DATA_DIR, ["lyrics", "track_id", "track_name", "release_date"])
df = filter.live(df)
df = filter.acoustic(df)
df = filter.remix(df)
df = filter.cover(df)
df = filter.smart_cover(df)
```

### English language (en)

```python
en_df = filter.language(df, "tracks/language_mapping_trh-0.8.json", white_list=["en"])
en_df = en_df.sample(5)
```

```python
def comparing_keywords_extractor(df: pd.DataFrame):
    for _, row in df.iterrows():
        spacy_keywords = keyword_extractor.spacy(row["lyrics"])
        keybert_keywords = keyword_extractor.keybert(row["lyrics"])
        textrank_keywords = keyword_extractor.textrank(row["lyrics"], threshold=0.05)
        rake_keywords = keyword_extractor.rake(row["lyrics"])
        yake_keywords = keyword_extractor.yake(row["lyrics"])

        # sort by value
        spacy_keywords = sorted(spacy_keywords.items(), key=lambda x: x[1], reverse=True)
        keybert_keywords = sorted(keybert_keywords.items(), key=lambda x: x[1], reverse=True)
        textrank_keywords = sorted(textrank_keywords.items(), key=lambda x: x[1], reverse=True)
        rake_keywords = sorted(rake_keywords.items(), key=lambda x: x[1], reverse=True)
        yake_keywords = sorted(yake_keywords.items(), key=lambda x: x[1], reverse=True)

        # print top 10 keywords 
        pprint(row["track_name"])
        pprint("Spacy keywords:")
        pprint(spacy_keywords[:10])
        pprint("Keybert keywords:")
        pprint(keybert_keywords[:10])
        pprint("Textrank keywords:")
        pprint(textrank_keywords[:10])
        pprint("Rake keywords:")
        pprint(rake_keywords[:10])
        pprint("Yake keywords:")
        pprint(yake_keywords[:10])
        pprint("--------------------------------")
```

```python
comparing_keywords_extractor(en_df)
```
